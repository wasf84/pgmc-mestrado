```python
# Importar a biblioteca PANDAS
import pandas as pd

csv_alto = ['chuva_2044037.csv',
            'chuva_2144000.csv',
            'chuva_2144005.csv',
            'chuva_2144006.csv',
            'chuva_2144023.csv',
            'chuva_2144072.csv',
            'chuva_2144073.csv',
            'cota_61075000.csv',
            'cota_61078000.csv',
            'cota_61078500.csv',
            'cota_61078600.csv',
            'cota_61106900.csv',
            'cota_61135000.csv',
            'cota_61140000.csv',
            'vazao_61061080.csv',
            'vazao_61065080.csv',
            'vazao_61075000.csv',
            'vazao_61078000.csv',
            'vazao_61078600.csv',
            'vazao_61106900.csv',
            'vazao_61135000.csv',
            'vazao_61140000.csv']

p_ajust_alto = p_alto + "planilhas_ajustadas\\"

for f in csv_alto:

        # Carrega o arquivo
        df = pd.read_csv(p_alto+f, sep='\t', header=0)

        # Renomeia as colunas 'ano', 'mes' e 'dia' para poder fazer o parse da data posteriormente
        df.rename(columns={'ano': 'year', 'mes': 'month', 'dia': 'day'}, inplace=True)

        # Criar uma coluna extra de datetime, combinando as 3 colunas
        df["data"] = pd.to_datetime(df[['year', 'month', 'day']])

        # Formatar a coluna de datetime como YYYY-mm-dd
        df["data"] = df["data"].dt.strftime("%Y-%m-%d")

        # Limpa o DataFrame
        df.drop(columns=['year', 'month', 'day'], inplace=True)

        df.set_index('data', inplace=True)

        # Salvar para o arquivo
        df.to_csv(p_ajust_alto+f, sep='\t')

# =================================================================== #

csv_baixo = ['chuva_1950006.csv',
             'chuva_2050001.csv',
             'vazao_61998080.csv',
             'vazao_62020080.csv']

p_ajust_baixo = p_baixo + "planilhas_ajustadas\\"

for f in csv_baixo:

        # Carrega o arquivo
        df = pd.read_csv(p_baixo+f, sep='\t', header=0)

        # Renomeia as colunas 'ano', 'mes' e 'dia' para poder fazer o parse da data posteriormente
        df.rename(columns={'ano': 'year', 'mes': 'month', 'dia': 'day'}, inplace=True)

        # Criar uma coluna extra de datetime, combinando as 3 colunas
        df["data"] = pd.to_datetime(df[['year', 'month', 'day']])

        # Formatar a coluna de datetime como YYYY-mm-dd
        df["data"] = df["data"].dt.strftime("%Y-%m-%d")

        # Limpa o DataFrame
        df.drop(columns=['year', 'month', 'day'], inplace=True)

        df.set_index('data', inplace=True)

        # Salvar para o arquivo
        df.to_csv(p_ajust_baixo+f, sep='\t')

# =================================================================== #

csv_medio = ['chuva_2046009.csv',
             'vazao_61730080.csv']

p_ajust_medio = p_medio + "planilhas_ajustadas\\"

for f in csv_medio:

        # Carrega o arquivo
        df = pd.read_csv(p_medio+f, sep='\t', header=0)

        # Renomeia as colunas 'ano', 'mes' e 'dia' para poder fazer o parse da data posteriormente
        df.rename(columns={'ano': 'year', 'mes': 'month', 'dia': 'day'}, inplace=True)

        # Criar uma coluna extra de datetime, combinando as 3 colunas
        df["data"] = pd.to_datetime(df[['year', 'month', 'day']])

        # Formatar a coluna de datetime como YYYY-mm-dd
        df["data"] = df["data"].dt.strftime("%Y-%m-%d")

        # Limpa o DataFrame
        df.drop(columns=['year', 'month', 'day'], inplace=True)

        df.set_index('data', inplace=True)

        # Salvar para o arquivo
        df.to_csv(p_ajust_medio+f, sep='\t')
```
